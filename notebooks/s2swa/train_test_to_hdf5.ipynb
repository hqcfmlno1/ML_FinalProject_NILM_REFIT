```python
import pandas as pd
import h5py
import numpy as np
from tqdm import tqdm
import os
```

```python
base_dir = os.path.dirname(os.path.dirname(os.getcwd()))
data_dir = os.path.join(base_dir, 'data', 'processed_data', 'House2_full_shifted.csv')
train_dir = os.path.join(base_dir, 'data', 'train', 'train_house2.csv')
test_dir = os.path.join(base_dir, 'data', 'test', 'test_house2.csv')
```

```python
train_h5_dir = os.path.join(base_dir, 'data', 'train', 'train_house2.h5')
test_h5_dir = os.path.join(base_dir, 'data', 'test', 'test_house2.h5')
```

```python
train_reader = pd.read_csv(train_dir, chunksize = 100000)
test_reader = pd.read_csv(test_dir, chunksize = 100000)
```

### Chú ý ###
Vì trong lúc chia dataset sử dụng chunksize của pandas t đã sử dụng append và vấn để header nên là các header bao gồm tên các cột được chèn vào cùng luôn nên phải bỏ đi các hàng đó

```python
for chunk in tqdm(train_reader):
    drop_id = chunk[chunk['Aggregate_t0']=='Aggregate_t0'].index
    chunk.drop(drop_id, inplace = True)
    chunk = chunk.astype(np.float32)
    chunk.to_hdf(train_h5_dir, key = 'train_house2', mode = 'a', format = 'table',
                 append = True, complib='blosc', complevel=3)
print('completed train')
for chunk in tqdm(test_reader):
    drop_id = chunk[chunk['Aggregate_t0']=='Aggregate_t0'].index
    chunk.drop(drop_id, inplace = True)
    chunk = chunk.astype(np.float32)
    chunk.to_hdf(test_h5_dir, key = 'test_house2', mode = 'a', format = 'table',
                 append = True, complib='blosc', complevel=3)
print('completed test')
```
