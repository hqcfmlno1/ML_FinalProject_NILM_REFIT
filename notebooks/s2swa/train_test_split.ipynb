```python
base_dir = os.path.dirname(os.path.dirname(os.getcwd()))
data_dir = os.path.join(base_dir, 'data', 'processed_data', 'House2_full_shifted.csv')
```

```python
reader = pd.read_csv(data_dir, chunksize = 100000)
```

```python
next(reader)
```

```python
train_dir = os.path.join(base_dir, 'data', 'train', 'train_house2.csv')
test_dir = os.path.join(base_dir, 'data', 'test', 'test_house2.csv')
```

```python
for chunk in tqdm(reader):
    chunk_train = chunk[((chunk['dom']-1)//7+1)%2==0]
    chunk_test = chunk[((chunk['dom']-1)//7+1)%2==1]
    chunk_train.drop(columns = ['dom'], inplace = True)
    chunk_test.drop(columns = ['dom'], inplace = True)
    chunk_train.to_csv(train_dir, mode = 'a', index = False)
    chunk_test.to_csv(test_dir, mode = 'a', index = False)
```
