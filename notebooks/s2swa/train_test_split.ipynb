```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

```python
base_dir = os.path.dirname(os.path.dirname(os.getcwd()))
data_dir = os.path.join(base_dir, 'data', 'processed_data', 'House2_full_shifted.csv')
h5_dir = os.path.join(base_dir, 'data', 'processed_data', 'House2_full_shifted.h5')
train_h5_dir = os.path.join(base_dir, 'data', 'train', 'train_house2.h5')
test_h5_dir = os.path.join(base_dir, 'data', 'test', 'test_house2.h5')
test_visualize_dir = os.path.join(base_dir, 'data', 'test', 'test_house2_for_visualize.npy')
train_npy_dir = os.path.join(base_dir, 'data', 'train', 'train_house2.npy')
test_npy_dir = os.path.join(base_dir, 'data', 'test', 'test_house2.npy')
```

```python
reader = pd.read_csv(data_dir, chunksize = 100000)
```

```python
for chunk in tqdm(reader):
    chunk.to_hdf(h5_dir, key = 'house2', mode = 'a', format = 'table',
                 append = True, complib='blosc', complevel=5)
```

```python
dataset = pd.read_hdf(h5_dir, key='house2')
dataset = dataset.astype('float16')
dataset = dataset.to_numpy()
print(dataset.shape)  
```
