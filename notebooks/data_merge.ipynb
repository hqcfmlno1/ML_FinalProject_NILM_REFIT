```python
import pandas as pd
import os
```

```python
base_dir = os.path.dirname(os.getcwd())
data_dir = os.path.join(base_dir, 'data', 'processed_data')
```

```python
df1 = pd.read_csv(os.path.join(data_dir, 'House2_part1.csv'), encoding = 'utf-8')
df2 = pd.read_csv(os.path.join(data_dir, 'House2_part2.csv'), encoding = 'utf-8')
df3 = pd.read_csv(os.path.join(data_dir, 'House2_part3.csv'), encoding = 'utf-8')
df4 = pd.read_csv(os.path.join(data_dir, 'House2_part4.csv'), encoding = 'utf-8')
df5 = pd.read_csv(os.path.join(data_dir, 'House2_part5.csv'), encoding = 'utf-8')
```

```python
df = pd.concat([df1, df2, df3, df4, df5], axis = 0)
df.shape
```

```python
df.to_csv(os.path.join(data_dir, 'House2_full.csv'), index = False)
```
